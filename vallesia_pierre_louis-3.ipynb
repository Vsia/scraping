```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.imdb.com/list/ls000096828/"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

```python
#getting titles
scraped_movies_names=soup.find_all("h3",class_="lister-item-header")
scraped_movies_names
```

```python

movies=[]
for movie in scraped_movies_names:
    movies.append(movie.get_text().replace("\n",""))
movies
```

```python
#scraping movie ratings
scraped_movies_rate = soup.find_all("div",class_="ipl-rating-star small")
scraped_movies_rate
```

```python
ratings=[]

for rating in scraped_movies_rate:
    rating= rating.get_text().replace("\n","")
    ratings.append(rating)
ratings

```

```python
#scraping year of release 
scraped_movies_year = soup.find_all("span",class_="lister-item-year text-muted unbold")
scraped_movies_year
```

```python
years=[]

for year in scraped_movies_year:
    year= year.get_text().replace("\n","").replace("(I)","").replace("(II)","").replace("TV Movie","").replace("TV Short","")
    years.append(year)
years
```

```python
#movie ranking
scraped_movies_rank = soup.find_all("span",class_="lister-item-index")
scraped_movies_rank
```

```python
ranks=[]

for rank in scraped_movies_rank:
    rank= rank.get_text().replace(".","")
    ranks.append(rank)
ranks
```

```python
#Store data
data = pd.DataFrame()
data[" Rank"]=ranks
data[" Name"]=movies
data[" Year"]=years 
data[" Rating"]=ratings
data.head()
```

```python
#saving as csv
data.to_csv("parsed_ranking.csv",index=False)
```
